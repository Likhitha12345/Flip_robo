```python
driver = webdriver.Chrome(r'C:/Users/rjhcl/Downloads/chromedriver_win32/chromedriver.exe')
driver.get("https://www.flipkart.com/")
```

```python
product_search = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input[@type = 'text']")
product_search.send_keys("sunglasses")

button = driver.find_element_by_xpath("//div[@class = 'col-12-12 _2oO9oE']//button[@class = 'L0Z3Pu']")
button.click()
```

```python
Brand = []
Product_Description= []
Price = []
Discount =[]

while True:
    
    time.sleep(3)
        
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//div[@class = '_2WkVRV']"):
        Brand.append(i.text)
        if len(Brand) == 100:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//a"):
        Product_Description.append(i.get_attribute('title'))
        if len(Product_Description) == 200:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_25b18c']//div[@class = '_30jeq3']"):
        Price.append(i.text) 
        if len(Price) == 100:
            break    
    for i in driver.find_elements_by_xpath("//div[@class= '_3Ay6Sb']/span"):
        Discount.append(i.text)
        if len(Discount) == 100:
            break  
            
    if len(Discount) == 100:
        break
        
    for j in driver.find_elements_by_xpath("//nav[@class ='yFHi8N']/a[@class = '_1LKTO3']"):
        next=[j.get_attribute("href")]
    
    time.sleep(3)
        
    for n in next:
        driver.get(n) 
```

```python
Product = []
for i in range(0,len(Product_Description)):
    if i == 0 or  i/2 == i//2:
        Product.append(Product_Description[i])
```

```python
df = pd.DataFrame({"Brand": Brand, 
                                  "Product_Description": Product, 
                                  "Price": Price , 
                                  "Discount": Discount})
df
```

```python
driver.get("https://www.myntra.com/shoes")
```

```python
price = driver.find_elements_by_xpath("//ul[@class ='price-list']//label[@class ='common-customCheckbox vertical-filters-label']/div[@class = 'common-checkboxIndicator']") 
price[1].click()
black = driver.find_elements_by_xpath("//li[@class ='colour-listItem']//label[@class ='common-customCheckbox']/div[@class = 'common-checkboxIndicator']") 
black[0].click()
```

```python
Brand,description,price = [], [], []
q = 0

while q<2:
    time.sleep(3)
    for i in driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h3[@class = 'product-brand']"):
        Brand.append(i.text)
    
    for i in driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h4[@class = 'product-product']"):
        description.append(i.text)
    
    for i in driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//div[@class = 'product-price']"):
        price.append(i.text)
    
    nxt = driver.find_element_by_xpath("//a[@rel = 'next']")
    nxt.click()
    q +=1

price= [i.split('Rs. ')[1] for i in price]   
```

```python
df = pd.DataFrame({"Brand":Brand,"description":description,"price":price})
df
```

```python
search = driver.find_element_by_xpath("//div[@class = 'nav-search-field ']/input[@type='text']")
search.send_keys("Laptop")
search_btn = driver.find_element_by_xpath("//input[@id= 'nav-search-submit-button']")
search_btn.click()
```

```python
filter_btn = driver.find_elements_by_xpath("//ul[@aria-labelledby= 'p_n_feature_thirteen_browse-bin-title']//div[@class= 'a-checkbox a-checkbox-fancy s-navigation-checkbox aok-float-left']")
filter_btn[-3].click()
```

```python
title = []
rating = []
price = []

for i in driver.find_elements_by_xpath("//span[@class= 'a-size-medium a-color-base a-text-normal']")[:10]:
    title.append(i.text)
for i in driver.find_elements_by_xpath("//span[@class = 'a-price-whole']")[:10]:
    price.append(i.text)  
```

```python
urls = []
for i in driver.find_elements_by_xpath("//a[@class = 'a-link-normal a-text-normal']"):
    urls.append(i.get_attribute("href"))
rating = []
for i in urls[:10]:
    driver.get(i)
    drop_down = driver.find_element_by_xpath("//i[@class = 'a-icon a-icon-popover']")
    drop_down.click()
    try:
        rating_ = driver.find_element_by_xpath("//span[@class = 'a-size-medium a-color-base a-text-beside-button a-text-bold']")
        rating.append(rating_.text)
    except:
        rating.append('-')
```

```python
df = pd.DataFrame({})
df['title'] = title
df['price'] = price
df['rating'] = rating
```
