```python
users_ratings = []
helpfull = []
review = []
summary = []
```

```python
link = driver.find_elements_by_xpath("//a[@class = '_1fQZEK']")
for i in link:
    i.click()
    time.sleep(5)
    for a in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
        users_ratings.append(a.text)
        time.sleep(3)
        time.sleep(3)
    for c in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div[1]/div[1]"):
        review.append(c.text)
        time.sleep(3)
    for d in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
        summary.append(d.text)
        time.sleep(3)
```
