```python
product_search = str(input("the product you want to search: "))
search_product = driver.find_element_by_xpath("//input[@class = 'nav-input nav-progressive-attribute']")
search_product.send_keys(product_search)
```

```python
search_btn = driver.find_element_by_xpath("//input[@id = 'nav-search-submit-button']")
search_btn.click()
```

```python
start_page = 0
end_page = 3
urls = []
for page in range(start_page,end_page):
    try:
        page_url = driver.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]')
        
        for url in page_url:
            url = url.get_attribute('href')     
            if url[0:4]=='http':                
                urls.append(url)    
        
        next_button = driver.find_element_by_xpath('//li[@class="a-last"]/a')      
        if next_button.text == 'Next→':                                            
            next_button.click()                                                    
            time.sleep(5)                                                            
        elif driver.find_element_by_xpath('//li[@class="a-disabled a-last"]/a').text == 'Next→':    
            print("No new pages exist.") 
            break
            
    except StaleElementReferenceException as e:             
        print("Stale Element Reference Exception")
        next_page = next_button.get_attribute('href')        
        driver.get(next_page)
```

```python
Brand_name = []
Product_name = []
Rating = []
No_of_ratings=[]
Price = []
Return_or_exchange = []
delivery = []
Availibility = []
Other_details = []
Product_url = []
```

```python
for url in urls:
    driver.get(url)                                                     
    print("Scraping URL = ", url)
    
    try:
        brand = driver.find_element_by_xpath('//a[@id="bylineInfo"]')      
        Brand_name.append(brand.text)
    except NoSuchElementException:
        Brand_name.append('-')
    
    try:
        name = driver.find_element_by_xpath('//h1[@id="title"]/span')      
        Product_name.append(name.text)
    except NoSuchElementException:
        Product_name.append('-')
    
    try:
        rating = driver.find_element_by_xpath('//span[@id="acrPopover"]')  
        Rating.append(rating.get_attribute("title"))
    except NoSuchElementException:
        Rating.append('-')
    
    try:
        n_ratings = driver.find_element_by_xpath('//a[@id="acrCustomerReviewLink"]/span')     
        No_of_ratings.append(n_ratings.text)
    except NoSuchElementException:
        No_of_ratings.append('-')
    
    try:
        product_price = driver.find_element_by_xpath('//span[@id="priceblock_ourprice"]')  
        Price.append(product_price.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:                                                                                     
        return_ = driver.find_element_by_xpath('//div[@data-name="RETURNS_POLICY"]/span/div[2]/a')
        Return_or_exchange.append(return_.text)
    except NoSuchElementException:
        Return_or_exchange.append('-')
    try:
        dele = driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]/b')         
        delivery.append(dele.text)
    except NoSuchElementException:
        delivery.append('-')
    
    try:
        avail = driver.find_element_by_xpath('//div[@id="availability"]/span')        
        Availibility.append(avail.text)
    except NoSuchElementException:
        Availibility.append('-')
    
    try:                                                                                    
        details = driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')
        Other_details.append('  ||  '.join(details.text.split('\n')))
    except NoSuchElementException:
        Other_details.append('-')
    
    Product_url.append(url)                                                            
    time.sleep(2)
```

```python
df = pd.DataFrame({})
df["Brand Name"] = Brand_name
df["product name"] = Product_name
df["Rating"] = Rating
df["no of ratings"] = No_of_ratings
df["price"] = Price
df["return/exchange"] = Return_or_exchange
df["expected delivery"] = delivery
df["Availability"] = Availibility
df["Other details"] = Other_details
df["Product url"] = Product_url
```
